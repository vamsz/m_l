```python
#import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
#read csv file
movie = pd.read_csv("~/Downloads/Data Science/data set/my_movies.csv")
movie
```

```python
movie.drop(['V1','V2','V3','V4','V5'], axis=1, inplace=True)
```

```python
# Building the model 
movie_apiori = apriori(movie, min_support = 0.1, use_colnames = True) 
```

```python
# Collecting the inferred rules in a dataframe 
rules = association_rules(movie_apiori, metric ="lift") 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 
print(rules.head())
```
