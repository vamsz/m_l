```python
#import libabry

import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
X = pd.read_csv("~/Downloads/Data Science/data set/SalaryData_Train.csv")
test = pd.read_csv("~/Downloads/Data Science/data set/SalaryData_Test.csv")
```

```python
X
```

```python
categorical = [col for col in X.columns if X[col].dtype==object]
categorical1 = [col for col in test.columns if test[col].dtype==object]
```

```python
for col in categorical:
    temp = {}
    count = 0
    for val in X[col].values:
        try:
            temp[val]
        except:
            temp[val] = count
            count += 1
    for val in test[col].values:
        try:
            temp[val]
        except:
            temp[val] = count
            count += 1
    X[col] = [temp[x] for x in X[col].values]
    test[col] = [temp[x] for x in test[col].values]
print(X[categorical])
```

```python
#spilit data

ytrain = X.Salary
X.drop(['Salary'], axis=1, inplace=True)
Xtrain = X
del X

ytest = test.Salary
test.drop(['Salary'], axis=1, inplace=True)
Xtest = test
del test
```

```python
# implement Model
ignb = GaussianNB()
pred_gnb = ignb.fit(Xtrain,ytrain).predict(Xtest)

#multinomial naive bayes

imnb = MultinomialNB()
pred_mnb = imnb.fit(Xtrain,ytrain).predict(Xtest)
```

```python
confusion_matrix(ytest,pred_gnb) # GaussianNB model
pd.crosstab(ytest.values.flatten(),pred_gnb) # confusion matrix using 
np.mean(pred_gnb==ytest.values.flatten()) # 100%
```

```python
# Confusion matrix multinomial naive bayes model
confusion_matrix(ytest,pred_mnb) # multinomial naive bayes model
pd.crosstab(ytest.values.flatten(),pred_mnb) # confusion matrix using 
np.mean(pred_mnb==ytest.values.flatten()) # 60%
```
